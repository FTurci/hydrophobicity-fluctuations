```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import sem
```

```python
df = pd.read_csv("output.csv")
```

```python
# df.plot('step','pe')
```

```python
df.plot('step','temperature', legend=False)
# target_rho = 0.96
# plt.plot(df['step'], np.ones_like(df['step'])*target_rho, color='darkorange')
```

```python
df.plot('step','density', ylabel=r"$\rho~[g/cm^3]$", color='k', legend=False)
# plt.plot(df['step'], df['density'])
target_rho = 0.957
plt.plot(df['step'], np.ones_like(df['step'])*target_rho, color='darkorange')

# plt.ylim(0.75,0.9)
my_rho = df.loc[df['step']>10000,'density'].mean()
print(my_rho, (my_rho-target_rho)/target_rho)
```

```python
df.plot('step', 'pressure')
plt.ylim(-2000,2000)
```

```python
pressure = df.loc[df['step']>1000,'pressure']
print(len(pressure),"values")
stride = 1
print("pressure", pressure[::stride].mean(), sem(pressure[::stride]), np.std(pressure[::stride]))
```

```python
density = df.loc[df['step']>10000,'density']
print("density", density.mean(), sem(density))
```
